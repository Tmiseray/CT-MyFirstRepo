```python
# Exercise 1: Movie Night Decision: 

# You are devloping a feature for a movie streaming platform that suggests 
# a movie genre to users based on their current mood and the day's weather. 

# The platform had the following recommendation criteria:
# If the user is feeling "happy" and the weather is "sunny", recommend a "comedy"
# If the user is feeling "happy but the weather is not "sunny", recommend a "romantic"
# if the user is feeling "sad", recommend a "drama"
# For any other mood, recommend an "adventure"

# Write a Python program that:
# Asks the user about their current mood (happy/sad/adventurous)
# Asks the user about the day's weather (sunny/rainy)
# Determines the movie genre based on the above criteria
# Displays the recommended genre to the user

# Hint: Use nested if statements to determine the movie genre based on 
# the user's mood and the day's weather

input("""Hi there! 
      I will be helping you with movie suggestions today! 
      But first, I will need to ask you some questions to determine the best genre for you! 
      When you are ready, press Enter to continue.""")
user_mood = input("""First question:
                  What is your current mood? 
                  (happy/sad/adventurous)""")
weather = input("""Thank you!
                Next:
                What is the weather like today? 
                (sunny/rainy)""")

if user_mood == "happy":
    if weather == "sunny":
        print("I recommend a comedy!")
    elif weather != "sunny":
        print("Maybe a romantic movie would be good for you.")
elif user_mood == "sad":
    print("I think a good drama might be intriguing!")
else:
    print("Sounds like you're in for an adventure today!!")
```

```python
# Exercise 2: Quick Temperature Check: 

# You are programming a smart wardrobe assistant that suggests outfits to users based on 
# the day's temperature and the type of event they are attending. 

# The wardrobe has the following recommendation criteria:
# If the temperature is below 15C and the event is "formal", suggest a "warm formal suit"
# If the temperature is below 15C and the event is "casual", suggest a "cozy sweater and jeans"
# if the temperature is above 15C and the event is "formal", suggest a "light formal suit"
# For any other combination, suggest a "t-shirt and shorts"

# Write Python Program that:
# 1. Asks the user about the day's temperature in Celsius
# 2. Asks the user about the type of event they are attending (formal/casual)
# 3. Determines the outfit based on the above criteria
# 4. Displays the recommended outfit to the user

# Hint: Use nested if statements to determine the outfit based on the day's 
# temperature and the type of event

input("""Hey there fashionista!
      I am your personal wardrobe assistant!
      Here to help you with any and all outfit delimmas!
      Before I can suggest the perfect outfit, I will need to ask you a couple things.
      When you are ready,
      press Enter to begin!""")
temperature = float(input("""Great!
                    First question:
                    What is the temperature outside?
                    (Preferably in Celius)"""))
event_type = input("""Thank You!
                   Next question:
                   What type of event will you be attending?
                   (formal/casual)""")

if temperature < 15:
    if event_type == "formal":
        print("Based on your responses, I suggest a warm formal suit for your event.")
    elif event_type == "casual":
        print("Hmmm... In that case, just a cozy sweater and jeans should suffice!")
elif temperature >= 15:
    if event_type == "formal":
        print("I see... Due to weather, I would recommend a light formal suit.")
    else:
        print("Sounds like a T-shirt and shorts would be perfect!")
```
