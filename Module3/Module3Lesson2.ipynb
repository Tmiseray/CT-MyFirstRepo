```python
# Exercise 3:
# Video Solution:

def add_task(tasks):
    name = input("Enter employee's name: ")
    description = input("Enter task description: ")
    deadline = input("Enter deadline (dd/mm/yyyy): ")
    tasks.append((name, description, deadline))

def display_tasks(tasks):
    for task in tasks:
        name, description, deadline = task
        print(f"Employee: {name}, Task: {description}, Deadline: {deadline}")

def complete_task(tasks):
    task_to_remove = input("Enter the description of the completed task: ")
    for i, task in enumerate(tasks):
        if task[1] == task_to_remove:
            del tasks[i]
            print("Task completed and removed.")
            break
    else:
        print("Task not found.")


def main():
    tasks = []
    while True:
        print("\n1. Add a task")
        print("2. View tasks")
        print("3. Complete a task")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_task(tasks)
        elif choice == '2':
            display_tasks(tasks)
        elif choice == '3':
            complete_task(tasks)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 4: 
# Video Solution:

def calculate_average_temperature(weather_data):
    total_temp = sum(record[1]for record in weather_data)
    average_temp = total_temp / len(weather_data)
    print(f"Average Temperature: {average_temp:.2f}")

def find_highest_precipitation(weather_data):
    max_precipitation = max(weather_data, key=lambda record: record[2])
    print(f"Highest Precipitaiton on : {max_precipitation[0]}, Level: {max_precipitation[2]}")

def list_hot_days(weather_data):
    threshold = float(input("Enter temperature threshold: "))
    hot_days = [record for record in weather_data if record[1] > threshold]
    for day in hot_days:
        print(f"Date: {day[0]}, Temperature: {day[1]}, Precipitation: {day[2]}")


def main():
    weather_data = [
        ("01/01/2020", 25.5, 5),    # In Celsius
        ("02/01/2020", 26.7, 10),
        ("03/01/2020", 24.8, 7),
        # Additional weather data...
    ]
    while True:
        print("\n1. Calculate Average Temperature")
        print("2. Find Highest Precipitation")
        print("3. List Hot Days")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            calculate_average_temperature(weather_data)
        elif choice == '2':
            find_highest_precipitation(weather_data)
        elif choice == '3':
            list_hot_days(weather_data)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 5: 
# Video Solution:

def display_itineraries(itineraries):
    for itinerary in itineraries:
        traveler, flights = itinerary[0], itinerary[1:]
        print(f"\nTraveler: {traveler}")
        for flight in flights:
            print(f"  Flight: {flight[0]}, From: {flight[1]}, To: {flight[2]}")

def find_itinerary_by_traveler(itineraries, traveler_name):
    for itinerary in itineraries:
        if itinerary[0].lower() == traveler_name.lower():
            print(f"\nItinerary for {traveler_name}:")
            for flight in itinerary[1:]:
                print(f"  Flight: {flight[0]}, From: {flight[1]}, To: {flight[2]}")
                break
    else:
        print("No itinerary found for this traveler.")

def list_itineraries_from_city(itineraries, city):
    for itinerary in itineraries:
        for flight in itinerary[1:]:
            if flight[1].lower() == city.lower():
                print(f"Traveler: {itinerary[0]}, Flight: {flight[0]}, To: {flight[2]}")


def main():
    itineraries = [
        ("Alice", ("FL123", "New York", "London"), ("FL456", "London", "Paris")),
        ("Bob", ("FL789", "Tokyo", "San Francisco"), ("FL321", "San Francisco", "New York")),
        # Additional itineraries...
    ]

    while True:
        print("\n1. View all itineraries")
        print("2. Search itineraries by traveler")
        print("3. List itineraries from a city")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            display_itineraries(itineraries)
        elif choice == '2':
            find_itinerary_by_traveler(itineraries)
        elif choice == '3':
            list_itineraries_from_city(itineraries)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 8: Conference Speaker Schedule Organizer
"""
Objective: Design a Python program to organize and display a conference schedule. This exercise focuses on practicing tuple operations beyond joining, such as repetition, membership tests, and unpacking, to enhance skills in data structuring and presentation.

Problem Statement: You are creating a system for a tech conference to manage its schedule. The conference has multiple sessions, each represented as a tuple containing the session's title and speaker. Your task is to organize these sessions into a daily schedule and provide options to view the entire schedule, repeat sessions, and check if specific speakers are presenting.

**Instructions**:

- Create a tuple for each day's sessions.
- Implement a function to display the schedule for each day.
- Create a function to repeat a specific session on a day's schedule.
- Develop a function to check if a particular speaker is presenting on any day.
- Ensure the program allows users to select different functionalities (e.g., view schedule, repeat a session, check speaker, exit).
- Use tuple operations like repetition and membership tests, and unpack day schedules using unpacking.

**Hints**:

- When repeating a session, consider how tuple repetition can be used to create a new schedule.
- Use in for membership tests to check if a speaker is part of the schedule.
- Use any function to check if any speaker is part of the schedule.
- Unpack tuples when displaying the schedule for clarity.

"""
```

```python
# Exercise 8: 
# Video Solution:

def display_schedule(schedule):
    for day, sessions in schedule.items():
        print(f"\n{day}:")
        for session in sessions:
            print(f"  - Title: {session[0]}, Speaker: {session[1]}")

def repeat_session(schedule, day):
    session_title = input("Enter the title of the session to repeat: ")
    for session in schedule[day]:
        if session[0].lower() == session_title.lower():
            schedule[day] = schedule[day] + (session,)
            print(f"Session '{session_title}' repeated.")
            return
    print("Session not found.")

def check_speaker(schedule):
    speaker_name = input("Enter the speaker's name to check: ")
    for day, sessions in schedule.items():
        if any(speaker_name.lower() in session[1].lower() for session in sessions):
            print(f"{speaker_name} is presenting on {day}.")
            return
    print(f"{speaker_name} is not presenting.")


def main():
    schedule = {
        "Day 1": (("Opening Keynote", "Alice Johnson"), ("Python Workshop", "Bob Smith")),
        "Day 2": (("AI Panel", "Charlie Davis"), ("Closing Remoarks", "Alice Johnson")),
        # Additional days and sessions...
    }

    while True:
        print("\n1. View conference schedule")
        print("2. Repeat a session")
        print("3. Check if a speaker is presenting")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            display_schedule(schedule)
        elif choice == '2':
            day = input("Enter the day to repeat a session: ")
            if day in schedule:
                repeat_session(schedule, day)
            else:
                print("Invalid day.")
        elif choice == '3':
            check_speaker(schedule)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 9:
# Video Solution:

def list_stars_in_constellation(star_catalog, constellation_name):
    for star in star_catalog:
        name, constellation, _ = star
        if constellation.lower() == constellation_name.lower():
            print(f" - {name}")

def find_brightest_star(star_catalog):
    brightest_star = min(star_catalog, key=lambda star: star[2])
    print(f"Brightest Star: {brightest_star[0]}, Magnitude: {brightest_star[2]}")

def categorize_stars_by_magnitude(star_catalog):
    for star in star_catalog:
        name, _, magnitude = star
        category = "Bright" if magnitude < 3.0 else "Dim"
        print(f"{name}: {category}")


def main():
    star_catalog = [
        ("Sirius", "Canis Major", -1.46),
        ("Canopus", "Carina", -0.74),
        ("Arcturus", "Bootes", -0.05),
        # Additional stars...
    ]

    while True:
        print("\n1. List stars in a constellation")
        print("2. Find the birghtest star")
        print("3. Categorize stars by magnitude")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            constellation = input("Enter the constellation name: ")
            list_stars_in_constellation(star_catalog, constellation)
        elif choice == '2':
            find_brightest_star(star_catalog)
        elif choice == '3':
            categorize_stars_by_magnitude(star_catalog)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 10: Flight Route Optimization
"""
Objective: Develop a Python program to optimize flight routes for an airline. This exercise aims to illustrate the best practices and common pitfalls when working with tuples in Python, focusing on data integrity and efficient data handling.

Problem Statement: You are tasked with creating a system for an airline to manage and optimize its flight routes. Each route is represented as a tuple containing the origin, destination, and distance. The system should provide functionalities like verifying route integrity, finding the shortest route, and ensuring no duplicate routes are added.

**Instructions**:

- Store the airline's routes in a set of tuples to avoid duplicates.
- Implement a function to add a new route to the set, checking for duplicates.
- Create a function to verify the integrity of a route (e.g., origin and destination are not the same).
- Develop a function to Find the shortest route between two destinations.
- Ensure the program allows users to select different functionalities (e.g., add a route, verify route, find shortest route, exit).
- Utilize tuple properties and control structures for data validation and manipulation.

**Hints**:

- When adding routes, ensure the tuples are unique before adding them to the set.
- Use tuple unpacking and comparisons for route integrity checks.
- Finding the shortest route might involve filtering and sorting based on distance.

"""
```

```python
# Exercise 10: 
# Video Solution:

def add_route(routes):
    origin = input("Enter the origin: ")
    destination = input("Enter the destination: ")
    distance = float(input("Enter the distance: "))
    route = (origin, destination, distance)

    if route in routes:
        print("Route already exists.")
    elif origin == destination: # 'LA' == 'LA'
        print("Origin and destination cannot be the same.")
    else:
        routes.add(route)
        print("Route added successfully.")

def verify_route(routes):
    for route in routes:
        # ('LA', 'Atlanta', 1000.00)
        # ('LA', 'Atlanta', 1000.00)
        if route[0] == route[1]:
            print(f"Invalid route found: {route}")
            return
    print("All routes are valid.")

def find_shortest_route(routes, origin, destination):
    possible_routes = [route for route in routes if route[0] == origin and route[1] == destination]
    if not possible_routes:
        print("No routes found.")
        return
    
    shortest_route = min(possible_routes, key=lambda route: route[2])
    print(f"Shortest route: {shortest_route}")



def main():
    routes = set()

    while True:
        print("\n1. Add a flight route")
        print("2. Verify flight integrity")
        print("3. Find shortest route")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_route(routes)
        elif choice == '2':
            verify_route(routes)
        elif choice == '3':
            origin = input("Enter the origin: ")
            destination = input("Enter the destination: ")
            find_shortest_route(routes, origin, destination)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
